```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
general_csv = pd.read_csv('/home/lpazc/Documentos/Proyects/DroidMalDetect-Py/results/train_dataset_pca.csv', index_col=False)
```

```python
folder = "/home/lpazc/Documentos/Proyects/DroidMalDetect-Py/results/files_image/mean_variance"
import pandas as pd
for column in general_csv.columns:
    name =  column.replace("/", "-")
    mean = general_csv[column].mean()
    variance =  general_csv[column].var()
    plt.figure(figsize=(8, 6))
    values = [mean, variance]
    labels = ['Media', 'Varianza']
    plt.bar(labels, values, color=['b', 'r'], alpha=0.7)
    plt.title(f'Media y Varianza de {column}')
    plt.xlabel('Animales')
    plt.ylabel('Valor')
    plt.legend()
    plt.savefig(f'{folder}/{name}_plots.png', bbox_inches='tight')
    plt.close()
```
