```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
general_csv = pd.read_csv('/home/lpazc/Documentos/Proyects/DroidMalDetect-Py/results/general_malware.csv')
```

```python
general_csv.columns
```

```python
name_column = 'Bwd Header Length'
```

```python
plt.figure(figsize=(20, 10))
sns.kdeplot(data=general_csv[name_column], shade=True, color='red', bw_method = "silverman")
sns.kdeplot(data=general_csv["Fwd URG Flags"], shade=True, color='blue', bw_method = "silverman")
plt.xlabel('Eje X')
plt.ylabel('Densidad')
plt.title('Gr√°fico de Densidad')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.violinplot(general_csv["FIN Flag Count"], showmeans=False, showmedians=True)
plt.xlabel('Eje X')
plt.ylabel('Densidad')
plt.title('Gr√°fico de Densidad')
plt.legend()
plt.show()
```
